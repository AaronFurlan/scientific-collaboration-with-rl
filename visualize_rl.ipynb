```python
# 1. Load the aggregated data
file_path = "results/rl_summary_by_archetype_by_effort.parquet"
df = pd.read_parquet(file_path)
```

```python
# 2. Visualize the mean reward over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="step", y="mean_reward", hue="archetype", ci="sd")

plt.title("Comparison: RL_Agent vs. Heuristics (Reward: by_effort)")
plt.xlabel("Simulation Step")
plt.ylabel("Mean Accumulated Reward")
plt.legend(title="Agent Type", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()
```
